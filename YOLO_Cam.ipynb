```python
import cv2
from darkflow.net.build import TFNet
import numpy as np
import time
```

```python
options = { 'model': 'cfg/yolo.cfg',
            'load' : 'bin/yolov2.weights',
            'threshold ': 0.3
           }
```

```python
tfnet = TFNet(options)
```

```python
## create different color boxes ##
colors  = [tuple(255* np.random.rand(3)) for i in range(10)]
 
capture = cv2.VideoCapture(2)
capture.set(cv2.CAP_PROP_FRAME_WIDTH,720)
capture.set(cv2.CAP_PROP_FRAME_HEIGHT,480)

while True:
    stime = time.time()
    ret,frame = capture.read()
    results = tfnet.return_predict(frame)
    if ret :
        for color,result in zip(colors,results):
            t = (result['topleft']['x'],result['topleft']['y'])
            b = (result['bottomright']['x'],result['bottomright']['y'])
            l = (result['label'])
            c = (result['confidence'])
            text = '{}: {:.0f}'.format(l,c * 100)
            frame = cv2.rectangle(frame,t,b,color,5)
            frame = cv2.putText(frame,text,t,cv2.FONT_HERSHEY_COMPLEX,1,(0,0,0),2)
     
            
        cv2.imshow('frame',frame)
        print('FPS {:.1f}'.format(1/(time.time()-stime)))
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
capture.release()
cv2.destroyAllWindows()
        
```
